## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')
sns.set_context('paper')
%matplotlib inline
```

### The function is : f(x) = w^2 
Taking the derivatve w.r.t x,

d(f(x))/d(w) = 2*w

so the gradient descent algorithm will be,

**repeat until convergence {**

         w:= w - learning_rate*2*w

**}**

## Setting the value for w and Learning rate

```python
w = 100
learning_rate1 = 1
learning_rate2 = 0.1
learning_rate3 = 0.01
learning_rate4 = 0.001
learning_rate5 = 0.0001
```

## Gradient Descent Algorithm

```python
grad1 = []
for i in range(20):
    w = w - learning_rate5*2*w
    grad1.append(w)
iterations1 = np.arange(0,20,1)
```

```python
grad2 = []
for i in range(40):
    w = w - learning_rate4*2*w
    grad2.append(w)
iterations2 = np.arange(0,40,1)
```

```python
grad3 = []
for i in range(60):
    w = w - learning_rate3*2*w
    grad3.append(w)
iterations3 = np.arange(0,60,1)
```

```python
grad4 = []
for i in range(80):
    w = w - learning_rate2*2*w
    grad4.append(w)
iterations4 = np.arange(0,80,1)
```

```python
grad5 = []
for i in range(100):
    w = w - learning_rate1*2*w
    grad5.append(w)
iterations5 = np.arange(0,100,1)
```

## Plotting the Loss 

```python
plt.plot(iterations1,grad1)
plt.xlabel('Number of Iterations')
plt.ylabel('Cost')
plt.title('Gradient Descent Algorithm')
plt.show()
```

```python
plt.plot(iterations2,grad2)
plt.xlabel('Number of Iterations')
plt.ylabel('Cost')
plt.title('Gradient Descent Algorithm')
plt.show()
```

```python
plt.plot(iterations3,grad3)
plt.xlabel('Number of Iterations')
plt.ylabel('Cost')
plt.title('Gradient Descent Algorithm')
plt.show()
```

```python
plt.plot(iterations4,grad4)
plt.xlabel('Number of Iterations')
plt.ylabel('Cost')
plt.title('Gradient Descent Algorithm')
plt.show()
```

```python
plt.plot(iterations5,grad5)
plt.xlabel('Number of Iterations')
plt.ylabel('Cost')
plt.title('Gradient Descent Algorithm')
plt.show()
```
