## Importing the Dependencies 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')
sns.set_context('paper')
%matplotlib inline
```

## Loading the Data 

```python
N = 50
```

```python
X = np.linspace(0,10,N)
Y = 0.5*X + np.random.randn(N)
```

```python
print("X = " + str(X[:5]) + " .... " + str(X[48:]))
print("\n")
print("Y = " + str(Y[:5]) + " .... " + str(Y[48:]))
```

## Adding Outliers 

```python
Y[-1] += 30
Y[-2] += 30
```

## Plotting the Data 

```python
plt.scatter(X,Y,c='b')
plt.show()
```

## Adding the Bias Term 

```python
X = np.vstack([np.ones(N), X]).T
```

```python
print("After adding the bias term: \n" + str(X[:5]))
```

## Maximum Likelihood Solution ( Linear Regression Without Regularization )

```python
w_ml = np.linalg.solve(X.T.dot(X),X.T.dot(Y))
```

```python
y_hat_ml = X.dot(w_ml)
```

```python
plt.scatter(X[:,1],Y,c= 'b')
plt.scatter(X[:,1],y_hat_ml,c='g')
plt.show()
```

## L2 Regularization Solution 

```python
l2 = 1000
```

```python
w_map = np.linalg.solve(l2*np.eye(2)+X.T.dot(X),X.T.dot(Y))
```

```python
y_hat_map = X.dot(w_map.T)
```

```python
plt.scatter(X[:,1],Y,c= 'b')
plt.scatter(X[:,1],y_hat_map,c='r')
plt.show()
```

## Comparison 

```python
plt.scatter(X[:,1],Y,c= 'b')
plt.scatter(X[:,1],y_hat_ml,c='g', label = 'Maximum Likelihood')
plt.scatter(X[:,1],y_hat_map,c='r', label = 'L2 Regularization')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Maximum Likelihood vs. L2 Regularization')
plt.legend(loc=6)
plt.show()
```
