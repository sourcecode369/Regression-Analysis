```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
sns.set_context('notebook')
```

```python
X = []
Y = []
for line in open("data_poly.csv"):
    x, y = line.split(',')
    x = float(x)
    X.append([1, x, x*x])
    Y.append(float(y))
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
sns.regplot(X[:,1], Y)
```

```python
w = np.linalg.solve(np.dot(X.T,X),np.dot(X.T,Y))
y_hat = np.dot(X, w)
```

```python
plt.scatter(X[:,1],Y)
plt.plot(sorted(X[:,1]),sorted(y_hat))
plt.show
```

```python
d1 = (Y - y_hat)
d2 = (Y - np.mean(Y))
```

```python
r2 = 1 - (np.dot(d1,d1)/np.dot(d2,d2)) 
```
